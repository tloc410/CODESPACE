## 3. Using list as intermediate data

```python
import cv2
import numpy as np

# read a grayscale image (nd.array)
data = cv2.imread('nature.png', 0)

# convert from np.array to 2d list
data = data.tolist()
height = len(data)
width = len(data[0])
print(f"Height: {height} and width: {width}")

# convert from 2d list to np.array
data = np.array(data)

# save the image
cv2.imwrite('1.nature_2.png', data)
```

## 4. Create black and noise images

```python
# create a black image

import cv2
import numpy as np

height = 800
width = 500
image = [[0]*width for _ in range(height)]

cv2.imwrite('2.black_image.jpg', np.array(image))
```

```python
# create a noise image

import cv2
import numpy as np
import random

height = 800
width = 500
image = [[0]*width for _ in range(height)]

# set noise to each pixel; a random integer in [0, 2550]
for i in range(height):
    for j in range(width):
        image[i][j] = random.randint(0, 255)

cv2.imwrite('2.noise_image.jpg', np.array(image))
```

## 5. Coppy an image

+ For gray images

```python
# copy image using list
import cv2
import numpy as np

# read an image
image = cv2.imread('nature.png', 0)
height, width = image.shape

# create a 2D list
# copy_image = [[0]*width  
#               for _ in range(height)]
copy_image = np.zeros((height, width))
copy_image = copy_image.tolist()

# copy
for i in range(height):
    for j in range(width):
        copy_image[i][j] = image[i][j]

# # save an image from a list
cv2.imwrite('copy.jpg', 
            np.array(copy_image))
```

## 6. Change illuminations

```python
# Copy and increase image brightness (1)

import cv2
import numpy as np

# read an image
image = cv2.imread('nature.png', 0)
height, width = image.shape

# create a 2D list
copy_image = [[0]*width for _ in range(height)]

# copy and increase image brightness
value = 50
for i in range(height):
    for j in range(width):
        copy_image[i][j] = min(255, image[i][j] + value)

# save an image from a list
cv2.imwrite('3.nature_increase_50.jpg', np.array(copy_image))
```

```python
# Copy and decrease image brightness (1)

import cv2
import numpy as np

# read an image
image = cv2.imread('nature.png', 0)
height, width = image.shape

# create a 2D list
copy_image = [[0]*width for _ in range(height)]

# copy and increase image brightness
value = 50
for i in range(height):
    for j in range(width):
        copy_image[i][j] = max(0, image[i][j] - value)

# save an image from a list
cv2.imwrite('3.nature_decrease_50.jpg', np.array(copy_image))
```
