```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
co2_levels = pd.read_csv('co2_levels.csv')
co2_levels = co2_levels.ffill()
co2_levels.head()
```

```python
co2_levels['datestamp'] = pd.to_datetime(co2_levels['datestamp'])

# set datetamp column as index
co2_levels = co2_levels.set_index('datestamp')
```

```python
co2_levels
```

### Decomposition

```python
decomposition = sm.tsa.seasonal_decompose(co2_levels['co2'])
fig = decomposition.plot()
plt.show()
```

### Time-based Features

```python
co2_levels['year'] = co2_levels.index.year
co2_levels['month'] = co2_levels.index.month
co2_levels['day'] = co2_levels.index.day 

co2_levels.head()
```
