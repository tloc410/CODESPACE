```python
import pandas as pd
import numpy as np
df = pd.DataFrame([[1, 2, 3],
                   [4, 5, 6],
                   [7, 8, 9],
                   [np.nan, np.nan, np.nan]],
                  columns=['A', 'B', 'C'])
df
```

```python
df.agg(['sum', 'min'])
```

```python
df.agg({'A': ['sum', 'min'], 'B': ['sum', 'max', 'min']})
```

```python
df.agg(x=('A', 'max'), y = ('B', 'min'))
```

```python
df.agg("mean", axis = "columns")
```
