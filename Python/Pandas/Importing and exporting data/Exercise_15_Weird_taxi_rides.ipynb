> 2 How many taxi rides had more than eight passengers?

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Data/nyc_taxi_2019-01.csv',
            usecols=['passenger_count', 'trip_distance', 'total_amount', 'payment_type'])
```

```python
df
```

```python
# How many taxi rides had more than eight passengers?
df.loc[df.passenger_count > 8].passenger_count.count()
```

```python
#  How many taxi rides had zero passengers?
df.loc[df.passenger_count == 0].passenger_count.count()
```

```python
# How many taxi rides were paid for in cash and cost over $1,000?
df.loc[(df.payment_type == 1 )& (df.total_amount > 1000)]
```

```python
# How many rides cost less than $0?
df.loc[df.total_amount < 0]
```

```python
# How many rides traveled a below-average distance but cost an above-average amount?
df.loc[(df.trip_distance < df.trip_distance.mean()) & (df.total_amount > df.total_amount.mean()) ]
```
