>As we’ve seen, CSV is a very flexible format. Many files that you wouldn’t necessarily think of as being CSV files can be imported into pandas with read_csv, thanks to a huge number of parameters that you can assign.
>
>In this exercise, I want you to create a data frame from a file that you wouldn’t nor mally think of as CSV but that fits the format fine: the Unix passwd file. This file, which is standard on Unix and Linux systems, contains usernames and passwords. Over the years, it has evolved such that it no longer contains the actual passwords. Although MacOS is based on Unix, it doesn’t use the passwd file for most user logins

```python
import pandas as pd
import numpy as np
```

```python
# C:\Users\loctr\OneDrive - hcmut.edu.vn\AIO\CODESPACE\CODESPACE\Python\Pandas\data\linux-etc-passwd.txt
df = pd.read_csv("../data/linux-etc-passwd.txt",
                 delimiter=":",
                 comment = '#',
                 names=['username', 'password', 'userid', 'groupid', 'name', 'homedir', 'shell'],
                 index_col='username')
```
