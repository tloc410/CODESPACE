>SOLUTION

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

np.random.seed(0)

months = 'Sep Oct Nov Dec Jan Feb Mar Apr May Jun'.split()

s = Series(np.random.randint(40, 61, 10),
          index=months)

print(f'Point average: {s.mean()}')

print(s + (80 - s.mean()))
```

Newcomers to pandas often assume that a series index must be unique. After all, the
index in a Python string, list, or tuple is unique, as are the keys in a Python dictionary.

### EXERCISE 7 ■ Long, medium, and short taxi rides

In this exercise, we once again look at taxi data—but instead of the number of passen-
gers, we examine the distance (in miles) each taxi ride went. Once again, I’ll ask you

to create a series based on a single-column CSV file, taxi-distance.csv.
First, load the data into a series. Then modify the series (or create another series)
containing category names rather than numbers based on these criteria:
+ Short, <= 2 miles
+ Medium, > 2 miles but <= 10 miles
+ Long, > 10 miles

Calculate the number of rides in each category.

```python
s = pd.read_csv('./taxi-distance.csv', header = None).squeeze()
categories = s.astype(str)
categories[s <= 2] = 'Short'
categories[(s > 2) & (s <= 10)] = 'Median'
categories[s > 10] = 'Long'
categories.value_counts()
```

```python
pd.cut(s,
bins=[0, 2, 10, s.max()],
include_lowest=True,
labels=['short', 'medium', 'long']).value_counts()
```

Compare the mean and median trip distances. What does that tell you about
the distribution of the data?

```python
print(f"mean trip distance {s.mean()}")
print(f"median trip distance {s.median()}")
```

How many short, medium, and long trips were there for trips that had only one
passenger? Note that the data for passenger count and trip length is from the
same data set, meaning the indexes are the same.

```python
pd.cut(s,
bins=[s.min(), 2, 10, s.max()],
include_lowest=True,
labels=['short', 'medium', 'long']).value_counts()
```

What happens if we don't pass explicit intervals, and instead ask pd.cut to just create 3 bins, with bins=3?
