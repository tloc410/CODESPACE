```python
import pandas as pd
from pandas import Series
from pandas import DataFrame

nyc_temp = pd.read_csv('./data/nyc-temps.txt').squeeze()
a = []
for i in range(1391):
    a.append(i*3%24)
df = DataFrame({'temp': nyc_temp,
               'hour': a})
df.temp-=10
```

>Set all the odd temperatures to the mean of all the temperatures.

```python
df.loc[(df.temp % 2) == 0, 'temp'] = df.temp.mean()
df
```

> Set the even temperatures at hours 9 and 18 to 3

```python
df.loc[df.hour.isin([9,18]), 'temp'] = 3
df
```

>If the hour is odd, then set the temperature to 5


```python
df.loc[df.hour % 2 == 1, 'temp'] = 5
df
```
